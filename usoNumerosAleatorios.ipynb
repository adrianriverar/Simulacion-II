```python
## Ejercicio 1.

# a)

import random as rand

if round(rand.random()) == 1:
  print('Sol')
else:
  print('Águila')
```

```python
# b)

n = 1000      # Número de volados
volados = []

for i in range(n):
  if round(rand.random()) == 1:
    volados.append('Sol')
  else:
    volados.append('Águila')

## Impresora "CHULEADOR 3000" (Tarea 28 marzo 2023).

from math import ceil
from tabulate import tabulate

n_row = 100              # Número deseado de filas a imprimir por subtabla
n_col = ceil(n / n_row)  # Número de columnas necesarias para imprimir subtablas

## AQUÍ TA LA MAGIA ¡CARAY!
volados_print = []
headers = []
for i in range(n_row):
  volados_bucket = []
  for j in range(n_col):
    headers.extend(['i', '       '])
    index = j*n_row + i
    if index < n:
      volados_bucket.extend([index + 1, volados[index]])
    else:
      volados_bucket.extend([' ', ' '])
  volados_print.append(volados_bucket)

print(tabulate(volados_print, headers))
```

```python
# c)

print('La probabilidad de que saliera SOL en ' + str(n) + ' volados es del ' + '{:.4%}'.format(volados.count('Sol') / n) + '.')
```
