```python
## Generador Congruencial Lineal Multiplicativo.
## X_i = (a*X_(i-1) + b) mod m
## r_i = X_i / m

import numpy as np
from random import randint

## Parámetros propuestos en "Numerical Recipes" (Press, Teukolsky, Vetterling, Flannery - 2007).
#m = 2**32       # Periodo (si cumple Teorema Hull-Dobell)
#a = 1664525     # Multiplicador
#b = 1013904223  # Incremento

## Parámetros arbitrarios.
m = 64
a = 15
b = 21

n_num = 100                       # Cantidad deseada de números pseudoaleatorios a generar
r = np.array([randint(0, m-1)])   # Lista numpy para pseudoaleatorios generados. Ingresar inicialmente semilla aleatoria en [0,m-1]
#r = np.array([21, 43])            # Semilla arbitraria

for i in range(n_num-1):
  r = np.append(r, (a*r[i] + b) % m)  # Generador de pseudoaleatorios en [0,m-1]

#for i in range(n_num-2):
#  r = np.append(r, (a*r[i] + b*r[i+1]) % m)  #Generador Fibonacci de pseudoaleatorios.

#r = r / m      # Restrigiendo pseudoaleatorios en [0,1)
r = r / (m-1)  # ¿Restrigiendo pseudoaleatorios en [0,1] o arruinando método?
#print(r)
```

```python
## Impresora "CHULEADOR 3000" (Tarea 28 marzo 2023).

from math import ceil
from tabulate import tabulate

n_dec = 8                    # Cantidad requerida de decimales a imprimir
n_row = 20                   # Número deseado de filas a imprimir por subtabla
n_col = ceil(n_num / n_row)  # Número de columnas necesarias para imprimir subtablas

## AQUÍ TA LA MAGIA ¡CARAY!
r_print = []
headers = []
for i in range(n_row):
  r_bucket = []
  for j in range(n_col):
    headers.extend(['i', 'r_i'])
    index = j*n_row + i
    if index < n_num:
      r_bucket.extend([index + 1, round(r[index], n_dec)])
    else:
      r_bucket.extend([' ', ' '])
  r_print.append(r_bucket)

print(tabulate(r_print, headers))
```
